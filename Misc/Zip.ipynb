# File Handling

Modes of opening files. By default "r" is assumed.

- r: Opens the file in read-only mode.
- w: Opens in write-only mode. The pointer is placed at the beginning of the file and this will overwrite any existing file with the same name. It will create a new file if one with the same name doesn't exist.



Example lets create a txt file.

````python
with open("file.txt",'w') as f:
    f.write("Hello World!")
````

Lets try reading it!

````python
with open("file.txt",'r') as f:
    print(f.read())
````

# Downloading files

````python
import requests
import matplotlib.pyplot as plt
from PIL import Image
````

````python
link = 'https://i.insider.com/5df126b679d7570ad2044f3e?width=1100&format=jpeg&auto=webp'
````

````python
r = requests.get(link) # Requesting the data
cont = r.content # accesing the data
````

````python
with open("image.png",'wb') as f:
    f.write(cont)
image = Image.open("image.png")

plt.imshow(image)
````
