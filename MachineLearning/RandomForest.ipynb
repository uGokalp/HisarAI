# Bagging (Bootstrap Aggreation)

Suppose the dataset has $N$ data points.

With __Bagging__ we sample $N$ points with replacement from the original dataset.

Build estimators with each sample.

__Average__ the prediction of the estimators.

__It turns out, we can decrease variance while keeping the bias the same.__

# Random Subspace Method

Random Subspace method is essentially Bagging but for features instead. So, all predictors will see a differet subset of features, which will allow the model to have interesting interactions between variables.

This is a method used to reduce to correlation between estimators. Imagine having 40 predictors, that have seen the exact same data. 

Would you need all 40 of them or just one?

# Out of Bag Error (OOB)

One thing I left out from you is if you sample $N$ points with replacement from $N$ points, it turns out roughly $33\%$ won't be sampled.

What does it mean for us?
- We can you the "unsampled" data to evaluate the performance of the models since the model essentially hasn't seen that subset of the data.

__Note:__ You with caution when you are working with time series data.
